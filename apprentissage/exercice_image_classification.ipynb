### Construire un ConvNet pour MNIST

Avec Keras (API fonctionnelle), construisez un réseau de neurones convolutif pour classifier des images **28×28 en niveaux de gris** en **10 classes** :
- une entrée `Input(shape=(28, 28, 1))`,
- plusieurs couches `Conv2D` (activation ReLU),
- des couches de **max-pooling** pour réduire la taille spatiale,
- une couche de **global average pooling**,
- une couche `Dense(10, softmax)` en sortie.
Créez ensuite l’objet `Model(inputs, outputs)`.

```python
import keras
from keras import layers

inputs = keras.Input(shape=(28, 28, 1))
x = layers.Conv2D(filters=64, kernel_size=3, activation="relu")(inputs)
x = layers.MaxPooling2D(pool_size=2)(x)
x = layers.Conv2D(filters=128, kernel_size=3, activation="relu")(x)
x = layers.MaxPooling2D(pool_size=2)(x)
x = layers.Conv2D(filters=256, kernel_size=3, activation="relu")(x)
x = layers.GlobalAveragePooling2D()(x)
outputs = layers.Dense(10, activation="softmax")(x)
model = keras.Model(inputs=inputs, outputs=outputs)
```

Puis écrivez une fonction qui copie un intervalle d’images (par index dans le nom de fichier) depuis le dossier original vers chaque sous-ensemble (train/val/test).

```python
import os, shutil, pathlib

original_dir = pathlib.Path("train")
new_base_dir = pathlib.Path("dogs_vs_cats_small")

def make_subset(subset_name, start_index, end_index):
    for category in ("cat", "dog"):
        dir = new_base_dir / subset_name / category
        os.makedirs(dir)
        fnames = [f"{category}.{i}.jpg" for i in range(start_index, end_index)]
        for fname in fnames:
            shutil.copyfile(src=original_dir / fname, dst=dir / fname)

make_subset("train", start_index=0, end_index=1000)
make_subset("validation", start_index=1000, end_index=1500)
make_subset("test", start_index=1500, end_index=2500)
```

### Mettre en place une pipeline de data augmentation

1. Définissez une liste de couches d’augmentation (ex. flip horizontal, rotation, zoom).
2. Écrivez une fonction `data_augmentation(images, targets)` qui applique ces couches aux images.
3. Créez un dataset augmenté en appliquant cette fonction avec `.map(..., num_parallel_calls=...)`.
4. Ajoutez `.prefetch(tf.data.AUTOTUNE)` pour optimiser le pipeline.

```python
data_augmentation_layers = [
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.2),
]

def data_augmentation(images, targets):
    for layer in data_augmentation_layers:
        images = layer(images)
    return images, targets

augmented_train_dataset = train_dataset.map(
    data_augmentation, num_parallel_calls=8
)
augmented_train_dataset = augmented_train_dataset.prefetch(tf.data.AUTOTUNE)
```

### Le "Backbone" en Apprentissage Profond

Un **backbone** est la partie centrale et pré-entraînée d'un réseau de neurones convolutifs (CNN), typiquement utilisé en vision par ordinateur pour l'extraction de caractéristiques (features) pertinentes à partir d'images.

**Points Clés :**

*   **Extracteur de Caractéristiques** : Il apprend à identifier des motifs génériques (bords, textures, formes) sur de vastes jeux de données (ex: ImageNet).
*   **Transfer Learning** : Permet de réutiliser ces connaissances acquises pour des tâches différentes, réduisant ainsi le besoin de données massives et le temps d'entraînement.
*   **Gel (Freezing)** : Initialement, le backbone est souvent "gelé" (`trainable=False`) pour:
    *   Préserver les caractéristiques génériques apprises.
    *   Prévenir le surapprentissage sur des jeux de données plus petits.
    *   Accélérer l'entraînement.
*   **Fine-tuning** : Après l'entraînement de la "tête" du modèle, le backbone peut être "dégelé" et entraîné avec un très faible taux d'apprentissage. Cela permet d'ajuster légèrement ses poids pour une meilleure adaptation à la tâche spécifique, tout en conservant ses capacités générales.

### Extraire des caractéristiques (feature extraction)

Écrivez une fonction qui parcourt un dataset (images, labels) et :
1. applique le préprocesseur aux images,
2. calcule les features via le backbone pré-entraîné,
3. concatène toutes les features et tous les labels,
4. renvoie deux tableaux NumPy : `(features, labels)`.

Utilisez-la pour préparer train/val/test.

```python
def get_features_and_labels(dataset):
    all_features = []
    all_labels = []
    for images, labels in dataset:
        preprocessed_images = preprocessor(images)
        features = conv_base.predict(preprocessed_images, verbose=0)
        all_features.append(features)
        all_labels.append(labels)
    return np.concatenate(all_features), np.concatenate(all_labels)

train_features, train_labels = get_features_and_labels(train_dataset)
val_features, val_labels = get_features_and_labels(validation_dataset)
test_features, test_labels = get_features_and_labels(test_dataset)
```

### L'Extracteur Fixe (Fixed Feature Extractor)

Un **extracteur fixe** est une partie d'un modèle (souvent un **backbone** pré-entraîné) dont les poids sont **gelés** (`trainable=False`) et ne sont donc pas mis à jour pendant l'entraînement.

**Points Clés :**

*   **Rôle** : Il est utilisé pour extraire des caractéristiques pertinentes des données sans modifier ses propres capacités d'extraction.
*   **Gel (Freezing)** : Les gradients ne sont pas calculés à travers l'extracteur fixe, et ses poids restent inchangés.
*   **Avantages** :
    *   **Prévention du Surapprentissage** : Idéal pour les petits jeux de données en réduisant le nombre de paramètres à entraîner.
    *   **Rapidité d'entraînement** : L'entraînement est accéléré car une grande partie du réseau ne nécessite pas de mise à jour des poids.
    *   **Stabilité** : Les caractéristiques extraites sont cohérentes car le mécanisme d'extraction ne change pas.
